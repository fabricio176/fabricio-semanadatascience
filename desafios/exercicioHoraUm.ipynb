```python
url = 'dataset/Libertadores_Matches.csv'

data = pd.read_csv(url)

data
```

```python
# Lendo as 5 primeiras partidas

data.head(5)
```

```python
data.tail(5)
```

```python
df = pd.DataFrame(data)

#Gols do time da casa
home_goals = df.groupby('home_team')['home_goal'].sum()

#Gols do time de fora
away_goals = df.groupby('away_team')['away_goal'].sum()

total_goals = pd.concat([home_goals, away_goals], axis=1).fillna(0)

total_goals['total_goals'] = total_goals['home_goal'] + total_goals['away_goal']

total_goals
```
